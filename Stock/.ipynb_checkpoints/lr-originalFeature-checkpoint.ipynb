```python
import pandas as pd
```

```python
data = pd.read_csv("heheda.csv")
data.dropna(inplace = True)
```

```python
train = data.iloc[:int(len(data) * 0.7)]
train.drop(["green", "ma_break_all_avg", "ma_break_30_avg", "tmrw_high_open", "tmrw_succeed_all_avg"], axis = 1, inplace = True)
train["tmrw_red"] = train["tmrw_red"].astype(int)

test = data.iloc[int(len(data) * 0.7):]
test.drop(["green", "ma_break_all_avg", "ma_break_30_avg", "tmrw_high_open", "tmrw_succeed_all_avg"], axis = 1, inplace = True)
test["tmrw_red"] = test["tmrw_red"].astype(int)
```

```python
train
```

```python
train_x = train.iloc[:, 1:-1]
train_y = train.iloc[:, -1]

test_x = test.iloc[:, 1:-1]
test_y = test.iloc[:, -1]
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()
clf.fit(train_x, train_y)
score = clf.score(test_x, test_y)
score
```

```python
clf.predict_proba(test_x)[:, 1]
```

```python
from sklearn.metrics import f1_score, roc_auc_score

roc_auc_score(test_y, clf.predict_proba(test_x)[:, 1])
```
