# Data Stats

* what we will do? We want to do some statistics about the data. About tags, imports, importfroms. 
* just a reminder. What is high_voted? note number > 2 is high voted. 

```python
import pandas as pd

def sort_and_save(by, asending, df, filename):
    sortedframe = df.sort_values(by= by, ascending= asending)
    sortedframe.reset_index(drop=True, inplace=True)
    sortedframe.to_csv(filename, index= False)
    
    ###########
    if by == ["kernels", "te"]:
        import numpy as np
        import matplotlib.pyplot as plt
        ## https://jingyan.baidu.com/article/1974b2894bb369f4b1f774b1.html
        %matplotlib inline
        plt.figure(figsize=(100,5))
        print(sortedframe["kernels"].describe())
        part = sortedframe[(sortedframe["kernels"] > 100) & (sortedframe["kernels"] <= 3000)]
        x = part["kernels"]#[i for i in range(len(part["kernels"]))]
        y = part["te"]
        plt.plot(x,y,color='red',label='kernel and te')
        cols = df.columns.values.tolist()
        b_max = 0
        for a, b, c in zip(x, y, part[cols[0]]):
            if a > b_max: 
                b_max = a
            plt.text(a, b, c, ha='center', va='bottom', fontsize=20)
        plt.hlines(0.5, 0, b_max)
        #plt.bar(x, y)
        #plt.show()
    return sortedframe.head(20)

def sum_col(df, newcol, cols= []):
    df[newcol] = df.apply(lambda row: sum([row[a] for a in cols]), axis=1)
    return

def target_encoding(n1, n2):
    return float(n1) / (n1 + n2)

def add_target_encoding(df, col= "te", from_col= ["hv_num", "lv_num"]): 
    df[col] = df.apply(lambda row: target_encoding(row[from_col[0]], row[from_col[1]]), axis=1)
    return 

def load_data(data, col_names):
    tmp = {}
    for name in col_names:
        tmp[name] = []
    for item in data:
        tmp[col_names[0]].append(item)
        tmp[col_names[1]].append(data[item][0])
        tmp[col_names[2]].append(data[item][1])
    return pd.DataFrame(tmp)
```
