```python
from matplotlib import pyplot as plt
from sklearn import svm
import pandas as pd

def f_importances(coef, names, top=-1):
    imp = coef
    imp, names = zip(*sorted(list(zip(imp, names))))

    # Show all features
    if top == -1:
        top = len(names)
    %matplotlib inline
    plt.barh(range(top), imp[::-1][0:top], align='center')
    plt.yticks(range(top), names[::-1][0:top])
    plt.show()
    return
```

```python
datafile = "J:/EECS6414/process/Pipeline/Data_cleaning/kernel_datassource.csv"
data = pd.read_csv(datafile)
data.head()
```

```python
env_vars = [    "km_dataSources", 
    "km_Data", 
    "km_versions", 
    "km_dataset_discussion_avg", 
      "km_competition_size_avg", 
#     "km_Visualization", 
#       "km_tags_te_avg", 
#       "km_Other", 
#       "km_isNotebook",
#       "km_Hidden",
#       "km_Notebook",
#       "km_null",
#       "km_language", 
           ]

data1 = data.sample(100)
features = data1[env_vars]
labels = data1["high_voted"]
```

SVM reference: 

https://stackoverflow.com/questions/41592661/determining-the-most-contributing-features-for-svm-classifier-in-sklearn

```python
# whatever your features are called
features_names = env_vars
clf = svm.SVC(kernel='linear')
clf.fit(features, labels)

f_importances(abs(clf.coef_), features_names, top=10)
```
